# Algerian Forest Fires Dataset

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

In Linear Regression, the output value will be FWI (continous value)

- day
- month
- year : Year which it is 
- Temperature : temperature noon (temperature max) in Celsius degrees: 22 to 42
- RH : Relative Humidity in %: 21 to 90
- Ws : Wind speed in km/h: 6 to 29
- Rain : total day in mm: 0 to 16.8
- FFMC : Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
- DMC : Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
- DC : Drought Code (DC) index from the FWI system: 7 to 220.4
- ISI : Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
- BUI : Buildup Index (BUI) index from the FWI system: 1.1 to 68
- FWI : Fire Weather Index (FWI) Index: 0 to 31.1
- Classes : Classes: two classes, namely fire and not fire.

```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the dataset
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
# Checking the dataset information - non null count and datatype
dataset.info()
```

#### Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

3 rows are having nan, 122 is an empty line, 123 is heading for the second set of dataset and 168 is having class as NaN

Dataset 1: Bejaia Region <br>
Dataset 2: Sidi Bel Region

```python
## Adding new column with region
dataset.loc[:122,'Region'] = 0
dataset.loc[122:,'Region'] = 1
```

```python
## Creating a new dataframe
df = dataset
```

```python
## checking the df info
df.info()
```

```python
## Checking first 5 records
df.head()
```

```python
## Converting float to int for Region
df[['Region']] = df[['Region']].astype('int')
```

```python
## Null count in df
df.isnull().sum()
```

```python
## Removing the null values and resetting the index
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
## This line has the header of 2nd dataset
df.iloc[[122]]
```

```python
## Dropping the header pf 2nd dataset
df = df.drop(122).reset_index(drop=True)
```

```python
## Checing the row 122
df.iloc[[122]]
```

```python
# Checking the column names
df.columns
```

```python
## Fixing the spaces in column names and checking
df.columns = df.columns.str.strip()
df.columns
```

```python
## Checking the datatype of the df and converting to proper datatype
df.info()
```

```python
## Changing the columns that have numbers from str to int and checking
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype('int')
df.info()
```

```python
## Changing the other columns to float datatype and checking
objects = [features for features in df.columns if df[features].dtypes == 'str']
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype('float')

df.info()
```

```python
## For monthly analysis, we need df

## Getting fire and not fire from df
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire','fire')
```

```python
## Monthly fire analysis - Sidi Bel
df_temp = df.loc[df['Region'] == 1]
figure = plt.figure(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data = df_temp, x = 'month', hue = 'Classes')
plt.ylabel('Number of Fires', weight = 'bold')
plt.xlabel('Months', weight = 'bold')
plt.title("Fire Analysis of Sidi-Bel Region", weight = 'bold')
plt.show()
```

```python
## Monthly fire analysis - Bejai region
df_temp = df.loc[df['Region'] == 0]
figure = plt.figure(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data = df_temp, x = 'month', hue = 'Classes')
plt.ylabel('Number of Fires', weight = 'bold')
plt.xlabel('Months', weight = 'bold')
plt.title("Fire Analysis of Bejai Region", weight = 'bold')
plt.show()
```

```python
## Monthly fire analysis - Overall region
figure = plt.figure(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(data = df, x = 'month', hue = 'Classes')
plt.ylabel('Number of Fires', weight = 'bold')
plt.xlabel('Months', weight = 'bold')
plt.title("Fire Analysis of Overall Region", weight = 'bold')
plt.show()
```
