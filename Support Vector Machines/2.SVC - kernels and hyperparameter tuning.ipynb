### Different Kernals in SVC
- linear
- rbf
- polynomial
- sigmoid

#### Dataset

```python
## Importing the libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Creating a synthetic dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 1000, n_features= 2,
                        n_classes = 2, n_clusters_per_class = 2, n_redundant = 0)
```

```python
## Scatter plot of the dataset
sns.scatterplot(x = pd.DataFrame(X)[0], y= pd.DataFrame(X)[1], hue = y)
```

There seems to be overlap between classes

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 10)
```

```python
## Importing the model library and performance metrics
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

#### linear kernel in SVC

```python
## Assigning linear kernel to SVC
linear = SVC(kernel = 'linear')
```

```python
## Training the model
linear.fit(X_train, y_train)
```

```python
## Predicting the values
y_pred = linear.predict(X_test)
```

```python
## performance of SVC - linear kernel
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

Accuracy is 91%, need to improve

#### rbf kernel in SVC

```python
## Assigning rbf kernel to model
rbf = SVC(kernel = 'rbf')
```

```python
## Training the model
rbf.fit(X_train,y_train)
```

```python
## Predicting the values
y_pred1 = rbf.predict(X_test)
```

```python
## Performance of rbf kernel SVC model
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

Accuracy is 92%

#### polynomial kernel in SVC

```python
## Assigning poly kernel to SVC model
polynomial = SVC(kernel = 'poly')
```

```python
## Training the model
polynomial.fit(X_train,y_train)
```

```python
## Predicting the values
y_pred2 = polynomial.predict(X_test)
```

```python
## Performance of the model
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

Accuracy is 89%

#### sigmoid kernel in SVC

```python
## Assigning sigmoid to SVC
sigmoid = SVC(kernel = 'sigmoid')
```

```python
## Training the model
sigmoid.fit(X_train,y_train)
```

```python
## Predicting the values
y_pred3 = sigmoid.predict(X_test)
```

```python
## Performance of the model
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

Accuracy is 83%

rbf kernel, known as radial basis function kernel, performs better than rest of the models. If nothing is specified in SVC, by default, rbf kernel is selected.

### Hyperparameter tuning with SVC

```python
## Importing GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
## Defining parameters 
param_grid = {'C': [0.1,1,10,100,1000],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel':['rbf']}
```

```python
## Assigning the parameters to GridSearchCV
grid = GridSearchCV(SVC(), param_grid = param_grid, refit = True, cv= 5, verbose = 3)
```

```python
## Training the model
grid.fit(X_train, y_train)
```

```python
## The best parameters are as follows
grid.best_params_
```

```python
## Predicting the values
y_pred4 = grid.predict(X_test)
```

```python
## Peformance of the model
print(confusion_matrix(y_test, y_pred4))
print(classification_report(y_test,y_pred4))
```

Accuracy is 97% and better with hyperparameer tuning
