### Logistic Regression 

```python
## Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Importing library to get the dataset for analysis
from sklearn.datasets import make_classification
```

```python
## Create the dataset 
X,y = make_classification(n_samples=1000, n_features = 100, n_classes = 2, random_state = 42)
```

```python
## Importing library for train test split
from sklearn.model_selection import train_test_split
```

```python
## Splitting the data into train and test for independent and dependent features
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=15)
```

```python
## Importing library for logistic regression analysis
from sklearn.linear_model import LogisticRegression
```

```python
## Initializing Logistic regression
logistic = LogisticRegression()
```

```python
## Training the model
logistic.fit(X_train,y_train)
```

```python
## Predicting the model
y_pred = logistic.predict(X_test)
```

```python
## Predicting the model probability of 2 classes
logistic.predict_proba(X_test)
```

```python
## Importing the performance metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
## Finding the performance of the model
score = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'Accuracy :{score}\n')
print(f'Confusion matrix: \n{cm}\n')
print('Classification report : \n',classification_report(y_test,y_pred))
```

Accuracy of logistic regression model is 0.81. The model performance can be increased using hyperparameter tuning. 
There are 2 types of hyperparameter tuning along with cross validation.
- Grid Search CV
- Randomized Search CV

#### Grid Search CV

```python
## Initializing logistic regression
model = LogisticRegression()

## To find the best parameters from the details given below using grid search CV
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']

params = dict(penalty = penalty, C = c_values, solver = solver)
```

```python
## Importing the library for gris search CV
from sklearn.model_selection import GridSearchCV
```

```python
## Impoirting stratified k fol for cross validation
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
## Assigning parameter to grid search cv
grid = GridSearchCV(estimator = model,\
                    param_grid = params,\
                    scoring = 'accuracy',\
                    cv = cv, n_jobs = 1)
```

```python
## Training the model with grid search CV
grid.fit(X_train,y_train)
```

```python
## Best parameters from grid search CV
grid.best_params_
```

```python
## Best Score from grid search CV
grid.best_score_
```

```python
## Predicted values
y_pred = grid.predict(X_test)
```

```python
## Finding the performance of the model
score = accuracy_score(y_test,y_pred)
print(f"Accuracy : {score}\n")
print(f"Confusion Matrix :\n {confusion_matrix(y_test,y_pred)}\n")
print(f"Classification Report : \n{classification_report(y_test,y_pred)}")
```

Accuracy increased to 0.83, precision reduced while recall increased

#### Randomized Search CV

```python
## Importing the library for randomized search CV
from sklearn.model_selection import RandomizedSearchCV
```

```python
## Initializing the model
model = LogisticRegression()
```

```python
## Giving parameters to randomized seach CV
randomcv = RandomizedSearchCV(estimator = model, 
                              param_distributions = params, 
                              cv = 5,
                              scoring = 'accuracy')
```

```python
## Training the model
randomcv.fit(X_train,y_train)
```

```python
## Finding the best parameters from Randomized search CV
randomcv.best_params_
```

```python
## Finding the best score from Randomized search CV
randomcv.best_score_
```

```python
## Predicted values
y_pred = randomcv.predict(X_test)
```

```python
## Performance of the randomized search CV logistic regression model
score = accuracy_score(y_test,y_pred)
print(f"Accuracy : {score}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test,y_pred)}\n")
print(f"Classification Report : \n {classification_report(y_test,y_pred)}")
```

Accuracy is 0.85, precison is better than grid seach CV. Recall is same in both grid search cv and randomized search cv. 
Randomized search CV has less FN + FP compared to normal model training and grid search CV. 
Better to go with Randomized Search CV
